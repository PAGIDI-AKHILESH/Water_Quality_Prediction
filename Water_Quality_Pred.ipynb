```python
#date is in object, so let us change it to datetime
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

```python
df.info()
```

```python
df = df.sort_values(by=['id','date'])
df.head()
```

```python
df.columns
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.columns
```

```python
pollutants = ['NH4', 'BSK5', 'Suspended', 'O2', 'NO3', 'NO2', 'SO4',
       'PO4', 'CL']
```

```python
### Week 2
```

```python
#dropping the missing values - dropna()
df = df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
x = df[['id', 'year']]
y = df[pollutants]
```

```python
#encoding - onehotencoder - 22 stations - 1 - 1
X_encoded = pd.get_dummies(x, columns = ['id'], drop_first = True)
```

```python
from sklearn.model_selection import train_test_split

# Train, TEst and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size = 0.2, random_state = 52
)
```

```python
model = MultiOutputRegressor(RandomForestRegressor(n_estimators = 100, random_state = 52))
model.fit(X_train, y_train)
```

```python
#Evaluate model
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data: ")
for i, pollutant in enumerate(pollutants):
  print(f'{pollutant}: ')
  print('     MSE: ', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
  print('     R2: ', r2_score(y_test.iloc[:, i], y_pred[:, i]))
```

```python
station_id = '22' # this is replacing real station id
year_input = 2025
input_data = pd.DataFrame({'year': [year_input],'id_': [station_id]})
input_encoded = pd.get_dummies(input_data, columns = ['id_'])
#Aligning training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
  input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]

#predicting pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in year {year_input}:")
for pollutant, value in zip(pollutants, predicted_pollutants):
  print(f"  {pollutant}: {value:.2f}")
```
